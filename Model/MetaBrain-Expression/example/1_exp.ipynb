```python
import scanpy as sc
import numpy as np
import pandas as pd
```

```python
import magic
```

```python
ad=sc.read('./scrna_FC_processed.h5ad')
ad
```

```python
sc.pp.normalize_total(ad, target_sum = 1e5)
```

```python
sc.pp.log1p(ad, base = 2)
df_adata = pd.DataFrame(ad.X.todense(), columns = ad.var_names, index = ad.obs_names)
df_adata
```

```python
#MAGIC
```

```python
magic_op = magic.MAGIC()
df_adata_magic = magic_op.fit_transform(df_adata)
```

```python
df_adata_magic2 = df_adata_magic*100
df_adata_magic2 = df_adata_magic2.astype(int)
df_adata_magic2 = df_adata_magic2/100
```

```python
adata_magic2 = sc.AnnData(df_adata_magic2)
```

```python
df_adata_magic2
```

```python
y1=np.array(df_adata_magic2.T)
```

```python
anno=ad.obs
anno
```

```python
anno['cell_type']=anno['ct']
```

```python
argidx = anno.cell_type.values.argsort()

anno = anno.iloc[argidx,:]
y_pb1 = y1[:, argidx]
```

```python
y_pb1.shape
```

```python
cluster_name, cnt = np.unique(anno.cell_type.values, return_counts=True)
cumcnt = np.cumsum(cnt)[:-1]
```

```python
y_pb1 = np.split(y_pb1, cumcnt, axis=1)
y_pb1 = np.array([tmp.mean(-1) for tmp in y_pb1]).T
```

```python
import pandas as pd
exp=pd.DataFrame(y_pb1,index=ad.var_names,columns=cluster_name)
exp
```

```python
exp.to_csv('./exp.csv')
```
